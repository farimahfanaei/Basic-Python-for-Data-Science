# Grids

Grids are general types of plots that allow you to map plot types to rows and columns of a grid


```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sb.load_dataset('iris')
```

```python
iris.head()
```

```python
sb.pairplot(iris)
```

we can do the same pairplot with grid in order to have much more control on that 

```python
sb.PairGrid(iris)
```

As you can see they are empty and this is the difference between pairplot nd PairGrid where youcan have more control on the plots 


The way to use this PairGrid is to set this PairGrid equal to a Variable and then map the plots to that variable:

```python
iris_grid= sb.PairGrid(iris)
iris_grid.map(plt.scatter)
```

```python
iris_grid= sb.PairGrid(iris)
iris_grid.map_diag(sb.distplot)
iris_grid.map_upper(sb.kdeplot)
iris_grid.map_lower(plt.scatter)
```

Here to discus more about grid we load the tips Data set:

```python
tips_dataset= sb.load_dataset('tips')
tips_dataset.head()
```

# FacetGrid()


```python
#it is a little bit similar to using subplot in matplotlib where we specified rows and columns :

g=sb.FacetGrid(data=tips_dataset,col='day', row='smoker')
```

So as we can see again we have empty grids and we need to map different plot types to these grids :

```python
g=sb.FacetGrid(data=tips_dataset, col='day', row='smoker')
g.map(sb.distplot, 'total_bill')
```

```python
#if you are dealing with a plot which needs more than 1 argument you  can add it as the second argument in the map call:
g=sb.FacetGrid(data=tips_dataset, col='day', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
