```python
import seaborn as sb 
%matplotlib inline 
tips_dataset = sb.load_dataset('tips')
```

```python
sb.lmplot(x='total_bill',y='tip',data=tips_dataset, hue='time', markers=['o','v'])
#by adding hue we compare also the categorical values and with market you can specify the data points more precisely
```

```python
#you can seprate the categorical values into 2 plot by adding another col or you can see different variable by adding row:
sb.lmplot(x='total_bill',y='tip',data=tips_dataset, col='day',row='time' , aspect=0.6, height=5)
```

# Some Style:


```python
sb.set_style('whitegrid')
sb.countplot(x='day', data=tips_dataset)
```

you can change the size using the fig object in matplotlib so you can combine the usage of these 2 libraries:

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
sb.countplot(x='day', data=tips_dataset)
```
