# Seaborn

Seaborn is another plotting library built on top og matplotlib, to install it : 

        conda install seaborn


```python
import seaborn as sb
```

```python
%matplotlib inline
```

```python
#seaborn have several built in data sets one of them is tips:
tips= sb.load_dataset('tips')
tips.head()
```

to check the distribution of a variable in a dataset:
    

```python
sb.distplot(tips['tip'])
```

in this plot you can also see the KDE (Kernel Density Estimation) which can be removed from the plot by :

        sb.distplot(tips['tip'], kde=False)

```python
#you can change the number of bins in the histagram
sb.distplot(tips['tip'], kde=False, bins=25)
```

# Jointplot() : 

allows you to combine 2 different distribution plots

```python
sb.jointplot(x='total_bill',y='tip',data=tips, kind='reg') 
#kind can give you the possibility to change the default (scatter) plot inside the jint plot
```

```python
sb.jointplot(x='total_bill',y='tip',data=tips, kind='kde') 
#kde shows you the density of the points where match the most
```

# Pairplot()

it allows you to plot pairwise relationships (so that it will plot all the numerical values of a data set and compare them all with each other)


```python
sb.pairplot(tips)
```

```python
#for the Categorical variables we can highlight them in the pairplot by specifying them in the 'hue' Parameter : 
sb.pairplot(tips, hue='sex')
```

# Rugplot()

it is similar to the histagram but it only shows every single point of a variable with a dash bar:


```python
sb.rugplot(tips['tip'])
```

# Categorical Data Plots

barplot is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
sb.barplot(x='day',y='total_bill',data=tips)
```

```python
#countplot is the same only it does not ask you about the y axis as it is by default the number of the categorical variables:
sb.countplot(x='day', data=tips)
```

# boxplot and violinplot

boxplots and violinplots are used to shown the distribution of categorical data

```python
sb.boxplot(x='sex', y='total_bill', data=tips,palette='rainbow')
```

```python
sb.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

# Violinplot()

It shows the distribution of quantitative data across several levels of one (or more) categorical variablesbut unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sb.violinplot(x="sex", y="tip", data=tips,palette='rainbow')
```

```python
sb.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sb.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```

# stripplot and swarmplot

The stripplot will draw a scatterplot where one variable is categorical. 
The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they dont overlap.

```python
sb.stripplot(x="day", y="total_bill", data=tips)
```

```python
sb.stripplot(x="day", y="total_bill", data=tips,jitter=False)
```

```python
sb.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sb.swarmplot(x="day", y="total_bill", data=tips)
```
