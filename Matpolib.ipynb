 we have 2 way of plotting one is Functional the other is Object oriented way which is the recommended method 


# Functional Method:

```python
plt.plot(x,y)
```

```python
#you can also have xlable, ylable and Title :
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('The first plot...')
plt.plot(x,y)
```

putting multi plots on the same canvas you can give it the number of rows and columns and the plot number

 plt.subplot(nrows, ncols, plot_number)

```python
plt.subplot(1,2,1)
# this line is creating a cell and bellow we plot 
plt.plot(x, y, 'r--')
# More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

# Object Oriented Methods:


```python
fig_obj=plt.figure()

#this figure is a kind of empty canvasin the following you can add set of axes to figure

axes = fig_obj.add_axes([0.1, 0.1, 0.8, 0.8])
# left, bottom, width, height (it is a list in range 0 to 1)
 
# Plot on that set of axes
axes.plot(x, y, 'g*')
axes.set_xlabel('X Axis') # Notice the use of set_ to begin methods
axes.set_ylabel('y Axis')
axes.set_title('Second Plot Using OO method...')
```

```python
fig_obj=plt.figure()
axes1 = fig_obj.add_axes([0.1, 0.1, 0.8, 0.8])


axes1.plot(x, y, 'g*')
axes1.set_xlabel('X Axis') # Notice the use of set_ to begin methods
axes1.set_ylabel('y Axis')
axes1.set_title('Second Plot Using Object Oriented method...') 
axes2= fig_obj.add_axes([0.2,0.55,0.35,0.25])
axes2.plot(y,x,'b--')
```

# subplots()

This is another Object Oriented method for ploting but the diifenerence with .figurre() is that here we can have number of Rows and Columns

```python
fig_obj, axes=plt.subplots(nrows=3,ncols=3)
```

To deal with the Overlapping issue:


```python
fig_obj, axes=plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig_obj, axes=plt.subplots(nrows=1,ncols=2)
```

Here the fig_obj is our Figure object and the axes is an array of axes to plot on:

```python
fig_obj, axes=plt.subplots(nrows=1,ncols=2)
for ax in axes :
    ax.plot(x,y,'b*')
```

As axes here is an array we can index it :


```python
fig_obj, axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,3*x,'r')
axes[0].set_title('Left Plot')
axes[1].plot(x,y,'b')
axes[1].set_title('Right Plot')
```

Working with Figure size and DPI using the Figure Object (fig_obj):

figsize is a tuple of the width and height of the figure in inches

dpi is the dots-per-inch (pixel per inch).


```python
fig_obj, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'b-')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('SubPlot');
```

# Saving figures

Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF

```python
fig_obj.savefig("PlotTest.png", dpi=200)
```

# More examples: 

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, 2*x, x, x**2)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
